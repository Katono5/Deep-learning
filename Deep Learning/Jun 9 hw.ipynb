```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

class Drawing:
    def __init__(self, x):
        self.x = x
    def draw_plot(self):
        data = np.arange(1, self.x[3])
        plt.subplot(221)
        # 设置x, y轴范围
        plt.xlim(1, 3)
        plt.ylim(1, 10)
        plt.plot(data, data*data)

    def draw_boxplot(self):
        plt.subplot(222)
        # 设置饼状图label,数据格式,饼块之间的举例
        plt.pie(self.x, labels=list('ABCD'), autopct='%.0f', explode=[0, 0.05, 0.05, 0])

    def draw_scatter(self):
        plt.subplot(223)
        # 设置x ,y轴名称
        plt.xlabel('x axis')
        plt.ylabel('random y')
        # 设置散点图marker
        plt.scatter(np.arange(0, self.x[3]), np.random.rand(self.x[3]), marker='*')

    def draw_barplot(self):
        plt.subplot(224)
        # 设置条形图purple
        plt.bar(self.x, self.x, color='purple')

test_Drawing = Drawing([1, 2, 3, 4])
test_Drawing.draw_boxplot()
test_Drawing.draw_barplot()
test_Drawing.draw_scatter()
test_Drawing.draw_plot()
```

```python
###########################
# 以下节选自之前做的iris数据处理#
############################
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
iris_data = load_iris()
# print(iris_data)
iris = pd.DataFrame(data=np.c_[iris_data['data'], iris_data['target']],
                    columns=iris_data['feature_names'] + ['species'])
def draw_pic(pic_kind, x_axis, y_axis):
    fig = iris[iris.species == 0].plot(kind=pic_kind, x=x_axis, y=y_axis,
                                       color='red', label='Setosa')
    iris[iris.species == 1].plot(kind=pic_kind, x=x_axis, y=y_axis,
                                 color='purple', label='versicolor', ax=fig)
    iris[iris.species == 2].plot(kind=pic_kind, x=x_axis, y=y_axis,
                                 color='green', label='virginica', ax=fig)
    # 规定两坐标轴名称
    fig.set_xlabel(str(x_axis))
    fig.set_ylabel(str(y_axis))
    fig.set_title(f'Relationship between {x_axis} and {y_axis}')
    # 展示图像
    plt.show()


def draw_boxplot(species):
    if species == 'setosa':
        specie = iris.loc[:49]
    elif species == 'versicolor':
        specie = iris.loc[50:99]
    elif species == 'virginica':
        specie = iris.loc[100:150]
    del specie['species']
    plt.boxplot(specie, vert=True, notch=True, patch_artist=True,
                boxprops={'color': 'orangered', 'facecolor': 'forestgreen'})
    plt.title(f"Features of the {species}")
    plt.xticks([1, 2, 3, 4], ['sepal length', 'sepal width', 'petal length', 'petal width'])
    plt.xlabel('Features')
    plt.ylabel("Length/Width (cm)")
    plt.show()
draw_boxplot('setosa')

draw_pic('scatter', 'sepal length (cm)', 'sepal width (cm)')
draw_pic('scatter', 'sepal length (cm)', 'petal width (cm)')
draw_pic('kde', 'sepal length (cm)', 'petal width (cm)')
draw_pic('kde', 'sepal width (cm)', 'petal length (cm)')
```
